# Analisa Data Kualifikasi Guru Provinsi 2012

# 1. Deskripsi Data

```python
import pandas as pd
data_guru = pd.read_csv('processed-data-guru-kualifikasi-provinsi-2012.csv', delimiter = ',', encoding = "ISO-8859-1")
shape = data_guru.shape
tipe = data_guru.dtypes
kolom = data_guru.columns
print("Dimensi Data :", shape)
print(kolom)
print("Type Data :")
print(tipe)
```

Dataset ini berisi mengenai jumlah persebaran guru per provinsi di Indonesia berdasarkan kualifikasi pada tahun 2012.
Format file pada dataset berikut adalah .csv, dengan infomasi dimensi 408 x 5

Penjelasan mengenai Variabel pada Dataset ini:

- kode_provinsi: Kode Provinsi
- nama_provinsi: Nama Provinsi
- tingkat_pendidikan: Tingkat Pendidikan
- kualifikasi_pendidikan_guru: Kualifikasi Pendidikan Guru
- jumlah_guru: Jumlah Guru

Type data per kolom pada data kualifikasi guru provinsi 2012 adalah :
- kode_provinsi             :     float
- nama_provinsi             :     string
- tingkat_pendidikan        :     string
- kualifikasi_pendidikan_guru  :   string
- jumlah_guru                  : float

```python
print("Berikut adalah ampel 5 random data kualifikasi guru provinsi 2012 :")
data_guru.sample(5)
```

# 2. Statistik Data

```python
sum = data_guru.jumlah_guru.sum()
min = data_guru.jumlah_guru.min()
max = data_guru.jumlah_guru.max()

print("Total Jumlah Guru :", sum)
print("Minimum Jumlah Guru :", sum)
print("Maximum Jumlah Guru :", sum)
```

```python
mean = data_guru.mean()
print("Rata-rata Jumlah Guru :", mean)
```

```python
median = data_guru.median()
print("Nilai Tengah dari Jumlah Guru :", median)
```

# 3. Visualisasi Data

```python
ax = data_guru.groupby('kualifikasi_pendidikan_guru').mean()['jumlah_guru']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Kualifikasi pendidikan", legend=True, fontsize=12)
ax.set_xlabel("Kualifikasi pendidikan", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_guru.groupby('tingkat_pendidikan').mean()['jumlah_guru']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tingkat pendidikan", legend=True, fontsize=12)
ax.set_xlabel("Tingkat Pendidikan", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_guru.groupby(['tingkat_pendidikan','kualifikasi_pendidikan_guru']).mean()['jumlah_guru']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tingkat pendidikan dan kualifikasi_pendidikan_guru", legend=True, fontsize=12)
ax.set_xlabel("Tingkat Pendidikan dan kualifikasi Pendidikan Guru", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_guru.groupby('nama_provinsi').mean()['jumlah_guru']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Provinsi", legend=True, fontsize=12)
ax.set_xlabel("Nama_Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

Bedasarkan kualifikasi pendidikan guru, jumlah rata-rata guru diatas atau sama dengan S1 memiliki nilai lebih tinggi dibandingkan dengan guru status pendidikannya dibawah S1.
Adapun nilai rata-rata tetinggi pada Tingkat pendidikan berada level Sekolah dasar. Hasil penggabungan dari variabel kualifikasi pendidikan guru dan tingkat pendidikan, nalai tertinggi pada masing-masing variabel tersebut masih berada pada nilai tinggi rata-rata dari keduanya dengan nilai rata-rata mendekati 5000 atau lebih tepatnya 48293.823529.
Dilihat dari Provinsi yang ada di Indonesi 2012, rata-rata jumlah guru terbanyak adalah pada Provinsi Nasional dengan nilai yang sangat signifikan dibandingkan dengan 33 provinsi lainnya. Prov. Nasional memiliki kualifikasi pendidikan guru diatas atau sama dengan S1 tertinggi pada tingkat pendidikan SD, kemudian disusul dengan pendidikan SMP, lalu SMA.
Namun pada kualifikasi pendidikan guru dibawah S1 Prov. Nasional, urutan nilai rata-rata tingkat pendidikan adalah SD, TK, SMP dan seterusnya sesuai pada tabel.

Berdasarkan hasil analisa data kualifikasi pendidikan guru tahun 2012, dapat disimpulkan bahwa guru sekolah di Indonesia dengan tingkat kualifikasi diatas atau sama dengan S1 sudah memiliki nilai yang cukup tinggi dan sebagian besar didominasi dengan pengajar sekolah Dasar, khususnya pada Prov. Nasional.
