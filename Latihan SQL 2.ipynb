```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationship

#   1. Report the account representative for each customer.

```python
run("""
SELECT 
    customerName as CustomerName,
    concat(firstName, lastName) as Employee
FROM
    customers c
        INNER JOIN
    employees e
    ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

# 2. Report total payments for Atelier graphique.

```python
run("""
select sum(amount) Total, customerName from payments p inner join customers c 
    ON p.customerNumber = c.customerNumber
    AND customerName ='Atelier graphique'""")
```

#  3. Report the total payments by date

```python
run("""
SELECT
    paymentDate as Date,
    sum(amount) as Total
FROM payments
GROUP BY paymentDate
""")
```

# 4. Report the products that have not been sold.

```python
run("""
SELECT
    p.productCode,
    p.productName,
    o.orderNumber,
    o.quantityOrdered
FROM products p
    LEFT JOIN 
    orderdetails o 
    ON p.productCode = o.productCode
WHERE o.productCode IS NULL
""")
```

# 5. List the amount paid by each customer.

```python
run("""
SELECT 
    customerName AS Customer_Name,
    sum(amount) AS Payment
FROM 
    payments p INNER JOIN customers c 
    using (customerNumber)
GROUP BY customerName
""")
```

# 6. How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT 
    customerName AS Customer_Name,
    p.`status` AS Status, 
    count(orderNumber) AS Total_order
FROM 
    orders p INNER JOIN customers c 
    ON p.customerNumber = c.customerNumber
    and status ='Shipped'
WHERE customerName ='Herkku Gifts'
    """)
```

# 7. Who are the employees in Boston?

```python
run("""
SELECT
    CONCAT(firstName, lastName) as Employee,
    city as City
FROM
    employees INNER JOIN offices
    USING (officeCode)
WHERE officeCode= 2
""")
```

# 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT
    customerName,
    sum(amount) as Total
FROM
    customers INNER JOIN payments
    USING (customerNumber)
GROUP BY customerNumber
HAVING sum(amount)>100000
ORDER BY sum(amount) DESC
""")
```

# 9. List the value of 'On Hold' orders.

```python
run("""
SELECT
    sum(quantityOrdered* priceEach) as Total,
    `status` AS Status
FROM 
    orders INNER JOIN orderdetails
    USING (orderNumber)
WHERE `status`='On Hold'
""")
```

# 10. Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT
    customerName AS Customer_Name,
    `status` AS Status,
    count(orderNumber) AS Total_Order
FROM 
    orders INNER JOIN customers
    Using (customerNumber)
WHERE `status`='On Hold'
GROUP BY customerNumber
""")
```

# Many to many relationship

# 1. List products sold by order date.

```python
run("""
SELECT
    productName,
    orderDate
FROM
    products p 
    INNER JOIN orderdetails o1
    ON p.productCode=o1.productCode
    INNER JOIN orders o2
    ON o1.orderNumber = o2.orderNumber
ORDER BY orderDate
""")
```

# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT
    productName AS Product_Name,
    orderDate AS Order_Date
FROM
    products p 
    INNER JOIN orderdetails o1
    ON p.productCode=o1.productCode
    INNER JOIN orders o2
    ON o1.orderNumber = o2.orderNumber
WHERE p.productName = '1940 Ford Pickup Truck'
ORDER BY orderDate DESC
""")
```

# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT
    c.customerName as Customer_Name,
    o2.orderNumber as Order_Number,
    sum(quantityOrdered*priceEach) as Total
FROM
    customers c 
    INNER JOIN orders o1
    ON c.customerNumber=o1.customerNumber
    INNER JOIN orderdetails o2
    ON o1.orderNumber = o2.orderNumber
GROUP BY o1.orderNumber
HAVING Total>25000
""")
```

# 4. Are there any products that appear on all orders?

```python
run("""
SELECT
    p.productName, 
    p.productCode,
    o1.orderNumber AS OrderNumber
FROM
    products p 
    LEFT JOIN orderdetails o1
    ON p.productCode=o1.productCode
    LEFT JOIN orders o2
    ON o1.orderNumber = o2.orderNumber
WHERE p.productCode in(o2.orderNumber) is not null
ORDER BY p.productName
""")
```

# 5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
select p.productName, (0.8*MSRP)
    from products p left join orderdetails o 
    ON p.productCode=o.productCode
WHERE (0.8*MSRP)< o.priceEach
""")
```

# 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
select *
    from products inner join orderdetails
    USING (productCode)
WHERE priceEach>=2* buyPrice
""")
```

# 7. What is the quantity on hand for products listed on 'On Hold' orders?

```python
#Each Product which is status = On Hold
run("""
SELECT
    p.productName,
    o2.`status`,
    count(o2.`status`) AS Total
FROM
    products p 
    INNER JOIN orderdetails o1
    ON p.productCode=o1.productCode
    INNER JOIN orders o2
    ON o1.orderNumber = o2.orderNumber
WHERE o2.`status`='On Hold'
GROUP BY p.productName
""")
```
