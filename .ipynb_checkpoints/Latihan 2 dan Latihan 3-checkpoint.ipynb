# LATIHAN 2 - Tugas List Nilai

```python
import random
import math

#GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1, 100,1))

print ("Nilai Tugas :")
print (nilaiTugas)

#Nilai ganjil
print("1.", "Angka Ganjil dari ke 250 nilai tersebut adalah")
Ganjil = []
for a in nilaiTugas:
    if(a%2!=0):
        Ganjil.append(a)
        print(a)

#Nilai > 70
More70 = []
for b in nilaiTugas:
    if (b>70):
        More70.append(b)
print("2.","Total yang mendapatkan nilai Lebih dari 70 adalah", len(More70))

#Standard Deviasi

rata = float (sum(nilaiTugas))/len(nilaiTugas)

dev = []

for c in nilaiTugas:
    dev.append(c - rata)
    
sqr = []
for c in dev:
    sqr.append(c * c)

mean = sum(sqr)/len(sqr)

standard_dev = math.sqrt(sum(sqr)/len(sqr)-1)
print("3.", "Standard Deviasi dari sekumpulan nilai tersebut adalah", standard_dev)

#NilaiMinMax
print("4.", "Nilai Minimum :", min(nilaiTugas),",", "Nilai maksimum:", max(nilaiTugas))

#Modus
freq = {}
for d in nilaiTugas:
    freq[d] = nilaiTugas.count(d)
    
ListSort = sorted(freq.items(), key=lambda kv: kv[1], reverse=True)
maxcount = ListSort[0][1]

modus = []

for d in ListSort:
    if d[1] == maxcount:
        modus.append(str(d[0]))
    else:
        break
print("5.", "Modus : ", ', '.join(modus),";", "Sebanyak :", maxcount)
```

# LATIHAN 3 - Word Count

```python
#Words count
paragraf = "Barcelona berhasil merebut posisi puncak klasemen La Liga Spanyol seusai menang 4-2 atas Sevilla pada pertandingan pekan kesembilan di Stadion Camp Nou, Sabtu (20/10/2018) atau Minggu dini hari WIB. Barcelona membuka keunggulan pada menit ke-2 melalui gol yang dicetak oleh Philippe Coutinho. Lionel Messi menggandakan keunggulan Barcelona, 10 menit berselang.  Namun, nahas bagi Barcelona, karena Messi harus ditarik keluar pada menit ke-26 setelah mengalami cedera. Tanpa Messi, skor 2-0 bertahan hingga babak pertama berakhir."
print(paragraf)
paragraf = paragraf.lower()
paragraf = paragraf.replace(',','')
paragraf = paragraf.replace('.','')
wordlist = paragraf.split()

freq = {}
for k in wordlist:
    freq[k] = wordlist.count(k)
    
for k, l in freq.items():
    print('-', k, ':', l)
```
